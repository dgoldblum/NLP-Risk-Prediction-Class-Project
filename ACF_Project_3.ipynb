```python
def optimizedLDA(df, outputdf):

  for col in df.columns:
    array = []
    documents = df[col].tolist()
    vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
    X = vectorizer.fit_transform(documents)
    model = LatentDirichletAllocation(n_components=10, random_state=42)
    dist = model.fit_transform(X)
    outputdf[col+'_lda'] = np.mean(dist, axis = 1).T
  return outputdf
```

Word2vec CBOW model

```python
def preprocessor(data):
  return preprocess_documents(data)

def word2vec(data):
  model = Word2Vec(sentences=data, vector_size=100, window=5, min_count=1, workers=4)
  embeddings = []
  for text in data:
      word_embeddings = [model.wv[word] for word in text if word in model.wv]
      if word_embeddings:
          paragraph_embedding = np.mean(word_embeddings, axis=0)
          embeddings.append(paragraph_embedding)

  return np.array(embeddings)

def get_w2v(df, outputdf):
  for col in df.columns:
    array = []
    preprocessed = preprocessor(df[col])
    data = word2vec(preprocessed)
    for i in range(len(data)):
      array.append(np.mean(data[i]))
    outputdf[col+'_w2v'] = array
    return outputdf
```

Doc2Vec

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from gensim.utils import simple_preprocess

def doc2vec(df, outputdf, epochs=10):
  for col in df.columns:
    array = []
    documents = df[col].tolist()
    tagged_data = [TaggedDocument(words=simple_preprocess(doc), tags=[str(i)]) for i, doc in enumerate(documents)]
    model = Doc2Vec(min_count=2, epochs=epochs)
    model.build_vocab(tagged_data)
    model.train(tagged_data, total_examples=model.corpus_count, epochs=model.epochs)
    array.append([model.infer_vector(doc.split()) for doc in documents])
    outputdf[col+'_d2v'] = np.mean(array, axis = 2).T
  return outputdf
```

TF-IDF

```python
def tfidf(df, outputdf):
  for col in df.columns:
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(df[col])
    outputdf[col+'_tfidf'] = np.mean(tfidf_matrix, axis=1)
  return outputdf
```

```python
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
bert_model = BertModel.from_pretrained(model_name)

def encode_sentences(sentences):
  input_ids = []
  attention_masks = []
  for sent in sentences:
    encoded_dict = tokenizer.encode_plus(sent, add_special_tokens=True, max_length=64, padding='max_length', return_attention_mask=True, return_tensors='pt')
    input_ids.append(encoded_dict['input_ids'])
    attention_masks.append(encoded_dict['attention_mask'])
  return torch.from_numpy(np.array(input_ids)), torch.from_numpy(np.array(attention_masks))


def get_senEmb(df, outputdf):
  for col in df.columns:
    array = []
    input_ids, attention_masks = encode_sentences(df[col])
    with torch.no_grad():
        outputs = bert_model(input_ids, attention_mask=attention_masks)
        sentence_embeddings = outputs[0][:, 0, :]
    for i in range(len(sentence_embeddings)):
      array.append(np.mean(sentence_embeddings[i]))
    outputdf[col] = array
```

```python
#LDA Engineered Feature
def textVect(texts):
  text_vectorizer = TextVectorization(max_tokens=1000, output_mode='int', standardize='lower_and_strip_punctuation',split='whitespace')
  text_vectorizer.adapt(texts)
  return text_vectorizer(texts)

def lda(tokens):
  lda_model = LatentDirichletAllocation(n_components=10, random_state=42)
  return lda_model.fit_transform(tokens)

def get_lda(df, outputdf):
  for col in df.columns:
    array = []
    preprocessed = textVect(df[col])
    data = lda(preprocessed)
    for i in range(len(data)):
      array.append(np.mean(data[i]))
    outputdf[col+'_lda'] = array
```

```python
# Use nn - probably a simple nn - few dense layers with a softmax/sigmoid output

input_shape = (xTrainC.shape[1],)

model = Sequential([
    Dense(50, activation='relu', input_shape=input_shape,),
    BatchNormalization(),
    Dropout(0.5),
    Dense(10, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid'),
])

# Compile the model

optimizer = tf.optimizers.Adam(learning_rate=0.007)

model.compile(optimizer=optimizer,
              loss='binary_focal_crossentropy',
              metrics=['accuracy'])
```
